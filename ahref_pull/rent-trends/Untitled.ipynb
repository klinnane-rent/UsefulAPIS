```python
df = pd.read_csv("rent.com-backlinks_history.csv")
```

```python
df['Referring Page URL'].apply(lambda x: x.split(".com")[0])
```

```python
count_referring_domains = pd.DataFrame(columns=['date', 'count'])
```

```python
df['First Seen'] = pd.to_datetime(df['First Seen'])
df['Day Lost']  =pd.to_datetime(df['Day Lost'])
```

```python
from datetime import datetime 
for idx, row in df.iterrows():
    first_seen_datetime = datetime(row['First Seen'].year, row['First Seen'].month, 1)
    if not count_referring_domains[count_referring_domains['date']  == first_seen_datetime].empty:
        my_index = count_referring_domains[count_referring_domains['date']  == first_seen_datetime].index[0]
        count_referring_domains.loc[my_index, 'count'] +=1 
    else:
        count_referring_domains.loc[len(count_referring_domains)] = [first_seen_datetime, 1]
    try:
        lost_seen_datetime = datetime(row['Day Lost'].year, row['Day Lost'].month, 1)
        if not count_referring_domains[count_referring_domains['date']  == lost_seen_datetime].empty:
            my_index = count_referring_domains[count_referring_domains['date']  == lost_seen_datetime].index[0]
            count_referring_domains.loc[my_index, 'count'] -=1 
        else:
            count_referring_domains.loc[len(count_referring_domains)] = [lost_seen_datetime, -1]
    except Exception as e:
        print('skipping. still live')
```
