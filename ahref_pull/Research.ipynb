```python
data = json.load(open('rent'))
df = pd.DataFrame(data['refdomains'])
df['target'] = 'rent'
appended_df = df
```

```python
domains = [ 'rentals', 'apartmentguide', 'rentcafe', 'apartments', 'forrent','zillow',  'hotpads', 'redfin', 'apartmentlist', 'realtor']

for domain in domains:
    data = json.load(open('new_lost_{}.json'.format(domain)))
    
    df = pd.DataFrame(data['refdomains'])
    df['target'] = domain
    appended_df = pd.concat([appended_df, df])
```

```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt



# Extract the 'clean' column
content_column = df["clean"]

# Initialize a Counter for word frequency analysis
word_counter = Counter()

# Initialize a list for sentence length analysis
sentence_lengths = []

# Process the content
for content in content_column:
    words = content.split()
    word_counter.update(words)
    sentence_lengths.extend([len(sentence.split()) for sentence in content.split('.')])

# Display word frequency analysis
common_words = word_counter.most_common(10)
print("Most common words:", common_words)

# Display sentence length analysis
plt.hist(sentence_lengths, bins=30)
plt.xlabel('Sentence Length')
plt.ylabel('Frequency')
plt.title('Sentence Length Distribution')
plt.show()

# Flag content with unusual sentence lengths
threshold = 30  # Set your threshold
flagged_content = df[df["clean"].apply(lambda x: any(len(sentence.split()) > threshold for sentence in x.split('.')))]

print("Content with unusual sentence lengths:")
print(flagged_content)
```

```python
appended_df.to_csv("new_lost_refdomains.csv")
```
